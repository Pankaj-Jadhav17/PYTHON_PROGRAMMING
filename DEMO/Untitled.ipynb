```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-15, 5, 200)

# Define the function
y = (x + 5)**2

# Plot
plt.plot(x, y, label="f(x) = (x + 5)^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graph of f(x) = (x + 5)^2")
plt.grid(True)
plt.legend()
plt.show()

```
